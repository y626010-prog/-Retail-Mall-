```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker()

# Customers
customers = pd.DataFrame({
    "customer_id": range(1,101),
    "first_name": [fake.first_name() for _ in range(100)],
    "last_name": [fake.last_name() for _ in range(100)],
    "email": [fake.email() if random.random()>0.05 else None for _ in range(100)],
    "gender": [random.choice(["M","F"]) if random.random()>0.05 else None for _ in range(100)],
    "age": [random.randint(18,65) for _ in range(100)],
    "loyalty_member": [random.choice([True,False]) for _ in range(100)],
    "phone_number": [fake.phone_number() if random.random()>0.07 else None for _ in range(100)]
})
print(customers.info())
print("__________________")
customers.head()
```

```python

# Branches
locations = ["Egypt", "UAE", "Saudi Arabia", "USA", "UK"]

branches = pd.DataFrame({
    "branch_id": range(1,6),
    "branch_name": [f"Mall_Branch_{i}" for i in range(1,6)],
    "location": [random.choice(locations) for _ in range(5)],
    "manager_name": [fake.name() if random.random()>0.05 else None for _ in range(5)],
    "open_since": [fake.date_between(start_date='-10y',end_date='-1y') for _ in range(5)]
})

print(branches.info())
print("__________________")
branches.head()
```

```python
fake = Faker()

brand_category = {
    "Electronics": ["Samsung","Apple","Sony","LG","HP","Lenovo"],
    "Clothing": ["Zara","H&M"],
    "Shoes": ["Nike","Adidas","Puma"],
    "Accessories": ["Rolex"],
    "Home": ["LG","Samsung"]
}

product_names = {
    "Electronics": ["Smart TV","Laptop","Smartphone","Tablet","Headphones","Bluetooth Speaker"],
    "Clothing": ["T-Shirt","Jeans","Jacket","Dress","Hoodie","Sweater"],
    "Shoes": ["Running Shoes","Sneakers","Boots","Sandals"],
    "Accessories": ["Watch","Sunglasses","Wallet","Backpack"],
    "Home": ["Microwave","Vacuum Cleaner","Coffee Maker","Blender"]
}

categories = list(brand_category.keys())

product_data = []

for i in range(1001,1101):

    category = random.choice(categories)
    brand = random.choice(brand_category[category])
    name = random.choice(product_names[category])

    product_name = f"{brand} {name}"

    product_data.append({
        "product_id": i,
        "product_name": product_name,
        "category": category,
        "brand": brand,
        "cost_price": round(random.uniform(50,500),2),
        "stock_quantity": random.randint(0,500)
    })

products = pd.DataFrame(product_data)
products.info()
products.head()
```

```python
# Sales
sales = pd.DataFrame({
    "sale_id": range(1,201),
    "customer_id": np.random.choice(customers["customer_id"],200),
    "branch_id": np.random.choice(branches["branch_id"],200),
    "sale_date": [fake.date_between(start_date='-1y',end_date='today') for _ in range(200)]
})
sales.info()
sales.head()
```

```python
# Sales Items
sales_items_list = []
item_id = 1
for sale in sales["sale_id"]:
    items_count = random.randint(1,4)
    for _ in range(items_count):
        product_id = random.choice(products["product_id"].tolist())
        cost = products.loc[products["product_id"]==product_id,"cost_price"].values[0]
        # unit_price منطقي أكبر من التكلفة
        unit_price = round(cost * random.uniform(1.2,1.8),2) if random.random()>0.02 else None
        quantity = random.randint(1,5) if random.random()>0.02 else None
        sales_items_list.append({
            "item_id": item_id,
            "sale_id": sale,
            "product_id": product_id,
            "quantity": quantity,
            "unit_price": unit_price
        })
        item_id += 1
sales_items = pd.DataFrame(sales_items_list)


sales_items.info()
sales_items.head()
```

```python
sales_items["total_amount"] = sales_items["unit_price"] *sales_items["quantity"]
```

```python
sale_totals = sales_items.groupby("sale_id")["total_amount"].sum().reset_index()
```

```python
sales_with_total = sales.merge(sale_totals, on="sale_id", how="left")
```

```python
sales_items.info()
```

```python
#payments
payments = pd.DataFrame({
    "payment_id": range(1, len(sales_with_total)+1),
    "sale_id": sales_with_total["sale_id"],
    "payment_status": [random.choice(["Completed","Pending","Failed"]) if random.random()>0.08 else None for _ in range(len(sales_with_total))],
    "payment_method": [random.choice(["Credit Card","Cash","Digital Wallet"]) for _ in range(len(sales_with_total))],
    "amount_paid": sales_with_total["total_amount"],
    "payment_date": sales_with_total["sale_date"]
})
payments.info()
payments.head()
```

```python

# Promotions
promotions = pd.DataFrame({
    "promo_id": range(1,11),
    "promo_name": [f"Season_Promo_{i}" for i in range(1,11)],
    "discount_percentage": [round(random.uniform(5,30),2) for _ in range(10)]
})

promotions.info()
promotions.head()
```

3. Home receives the highest average discount (27%), yet it's not the top-selling category, while Clothing has the lowest discount (21%) and also the lowest sales — confirming that discount percentage alone does not drive purchasing decisions for most categories.

*Instead of increasing discounts, focus on product visibility and targeted marketing to boost underperforming categories like Clothing.

Conclusion & Strategic Insights
Mall Management System - Branch Analysis: Mall_Branch

Based on a comprehensive data analysis of Mall_Branch_3 and a comparative study of financial performance and customer behavior, the following insights were established:
1. The Promotion Paradox & Profitability Analysis

    Financial Discrepancy: Analysis reveals that Branch 3 is currently the lowest-performing branch in terms of sales and net profit, despite maintaining average operational costs (Total Cost).

    The Impact of Discounts: A critical discovery shows that non-promotional transactions (No Promo) yield significantly higher profits than discounted ones.

    Peak vs. Low Performance: Thursday was identified as the peak day for both sales and profit, characterized by a minimal reliance on promotions. Conversely, Friday and Wednesday recorded the lowest profit margins, even though Friday featured the highest frequency of promotional offers.

    Insight: Current promotional strategies are "cannibalizing" profit margins without driving a proportional increase in sales volume. Selling at Full Price in this branch has proven to be more financially viable.

2. Category & Product Performance

    Lead Category: The Electronics category is the top-performing segment, accounting for the highest volume of transactions and serving as the branch's primary revenue driver.

    Growth Opportunities: While Thursday drives high-value category sales, Friday tends to deplete inventory through deep discounts without generating a satisfactory Return on Investment (ROI).

3. Customer Demographic Profiling

    Gender Dominance: Data confirms that Females represent the primary purchasing power, showing higher transaction frequency and spending compared to males.

    The Golden Age Group: The branch's core customer base is concentrated within the 18–25 age bracket, clearly defining the branch as a youth-oriented shopping destination.

Strategic Recommendations

    Inventory Optimization: Increase the stock variety and floor space for Female-oriented products and Electronics, while reducing the inventory allocation for lower-performing male-specific items.

    The "Thursday" Strategy: Capitalize on Thursday's peak performance by launching new product arrivals or "Value-based" offers instead of "Discount-based" ones to maintain high margins.

    Friday Promotion Adjustment: We recommend halting high direct discounts on Friday for high-demand categories. Promotions should be restricted only to slow-moving inventory (High Stock Quantity) to clear warehouse space.

    Targeted Marketing: Direct social media and marketing campaigns toward the 18–25 age group to sustain the flow of the branch's most active customer segment.
